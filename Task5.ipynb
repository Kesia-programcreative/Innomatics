```python
# Function to analyze post engagement
def analyze_engagement(likes_list):
    total_likes = 0
    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes
    # Determine post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    # Print results
    print("Total Likes:", total_likes)
    print("Post Status:", status)
# Example Data (likes from different platforms or time intervals)
likes_data = [200, 350, 150, 250, 100]
# Function Call
analyze_engagement(likes_data)
```

```python
# Function to monitor medicine stock
def check_medicine_stock(stock):
    # Conditional logic to check stock level
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"
    # Print results in required format
    print("Medicine Stock:", stock)
    print("Status:", status)
# Example usage
medicine_stock = 6
check_medicine_stock(medicine_stock)
```

```python
# Function to analyze rainfall data
def check_rainfall(rainfall_data, required_level):

    total_rainfall = 0

    # Loop to calculate total rainfall
    for rain in rainfall_data:
        total_rainfall += rain

    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_data)
    # Conditional logic to check adequacy
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    # Print results in required format
    print("Average Rainfall:", int(average_rainfall))
    print("Rainfall Status:", status)
# Example seasonal rainfall data (in mm)
season_rainfall = [60, 75, 80, 70, 75]
# Required rainfall level
required = 70
# Function Call
check_rainfall(season_rainfall, required)
```

```python
# Function to detect duplicate usernames
def detect_duplicate_accounts(usernames):
    # Convert list to set
    unique_usernames = set(usernames)
    # Check if duplicates exist
    if len(unique_usernames) < len(usernames):
        result = "Yes"
    else:
        result = "No"
    # Print result in required format
    print("Duplicate Accounts Found:", result)
# Example username list
user_list = ["alex01", "john_doe", "emma23", "alex01"]
# Function Call
detect_duplicate_accounts(user_list)
```

```python
# Function to check appointment eligibility
def check_eligibility(age):
    # Condition check
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"
    # Print result in required format
    print("Patient Age:", age)
    print("Eligibility Status:", status)
# Example usage
patient_age = 21
check_eligibility(patient_age)
```

```python
# Function to filter premium crop prices
def filter_premium_crops(prices):
    premium_crops = []
    # List filtering logic using loop
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
    # Print result in required format
    print("Premium Crops:", premium_crops)
# Example crop prices (in ₹)
crop_prices = [1500, 2500, 1800, 3200, 1900]
# Function Call
filter_premium_crops(crop_prices)
```

```python
# Function to evaluate application health
def check_application_health(error_count):
    # Conditional statements to determine system status
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    # Print results in required format
    print("Error Count:", error_count)
    print("System Status:", status)
# Example usage
errors = 7
check_application_health(errors)
```

```python
# Function to validate daily transaction amount
def check_transaction(amount):
    daily_limit = 50000  # ₹50,000 limit
    # Conditional logic
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    # Print result in required format
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)
# Example usage
transaction_amount = 60000
check_transaction(transaction_amount)
```

```python
# Function to calculate attendance percentage and check eligibility
def check_exam_eligibility(attendance_list):
    total_classes = len(attendance_list)
    attended_classes = 0
    # Loop to count attended classes
    for status in attendance_list:
        if status == "Present":
            attended_classes += 1
    # Calculate attendance percentage
    attendance_percentage = (attended_classes / total_classes) * 100
    # Check eligibility
    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"
    # Print result in required format
    print("Attendance Percentage:", round(attendance_percentage, 1))
    print("Exam Eligibility:", eligibility)
# Example attendance data
attendance_data = [
    "Present", "Present", "Absent", "Present", "Present",
    "Present", "Absent", "Present", "Present", "Present"
]
# Function Call
check_exam_eligibility(attendance_data)
```

```python
# Function to calculate electricity bill and usage classification
def calculate_electricity_bill(units):
    bill = 0
    # Slab calculation using conditional logic
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    # Usage Classification
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"
    # Return both values
    return bill, usage_status
# Example usage
units_consumed = 250
total_bill, status = calculate_electricity_bill(units_consumed)
print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Level:", status)
```
