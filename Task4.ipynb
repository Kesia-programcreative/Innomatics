```python
AVAILABLE_PLANS = [199, 299, 399, 599]
def check_recharge(amount):
    if amount < 50:
        print("Minimum recharge is ₹50")
        return False
    if amount not in AVAILABLE_PLANS:
        print("Invalid plan. Choose from:", AVAILABLE_PLANS)
        return False
    return True
while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))
        if check_recharge(amount):
            print("Recharge successful")
            break
        else:
            print("Try again")
    except ValueError:
        print("Enter numbers only")
```

```python
inventory = {
    "Laptop": 17,
    "Mouse": 12,
    "Keyboard": 18,
    "Monitor": 8
}
for product in inventory:
    quantity = inventory[product]
    if quantity < 15:
        print(product, ": Reorder Alert (Stock =", quantity, ")")
    else:
        print(product, ": Stock OK (Stock =", quantity, ")")
```

```python
def process_result(marks):
    average = sum(marks) / len(marks)
    if average >= 50:
        return "Pass", average
    else:
        return "Fail", average
student_marks = [40, 71, 55, 83]
result, avg = process_result(student_marks)
print("Average Marks:", avg)
print("Result:", result)
```

```python
def fare_calc(km, peak):
    fare = 50 + 12 * km
    if peak == "yes":
        fare *= 1.25
    return fare
while True:
    try:
        km = float(input("Distance in km: "))
        peak = input("Peak hour? (yes/no): ").lower()
        print("Total Fare: ₹", fare_calc(km, peak))
        break
    except ValueError:
        print("Enter valid numbers")
```

```python
def check_attendance(att_list):
    present = att_list.count("P")
    percentage = (present / len(att_list)) * 100
    if percentage >= 75:
        return "Eligible", percentage
    else:
        return "Not Eligible", percentage
attendance = ["P", "A", "P", "A", "A", "P", "P", "A"]
result, percent = check_attendance(attendance)
print("Attendance %:", percent)
print("Status:", result)
```

```python
#Password Strength Checker
def check_password(pwd):
    # password should be at least 8 characters
    if len(pwd) < 8:
        return "Weak Password"
    has_digit = False
    has_special = False
    for ch in pwd:
        # if number found
        if ch.isdigit():
            has_digit = True
        # if special character found
        if ch in "@#$":
            has_special = True
    # password is strong only if both conditions are true
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"
password = input("Enter password: ")
print(check_password(password))
```
