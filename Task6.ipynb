```python
def smart_parking(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            parked -= 1

        # Track peak usage
        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)
    print("Peak Parking Usage:", peak)

    if peak > capacity:
        print("Parking Status: Exceeded Capacity! Alert!")
    else:
        print("Parking Status: Available")


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking(capacity, vehicle_logs)
```

```python
def estimate_delivery_time(distance, traffic, weather):

    # Base time calculation
    base_time = distance * 2.5

    # Traffic delay
    if traffic == "Low":
        base_time += 5
    elif traffic == "Medium":
        base_time += 10
    elif traffic == "High":
        base_time += 20

    # Weather delay
    if weather == "Clear":
        base_time += 0
    elif weather == "Rainy":
        base_time += 15
    elif weather == "Stormy":
        base_time += 25

    print("Estimated Delivery Time:", int(base_time), "minutes")


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

estimate_delivery_time(distance, traffic, weather)
```

```python
def seat_occupancy(total_seats, booked_seats):

    # Count booked seats
    booked_count = len(booked_seats)

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 70:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

    # Suggest additional show
    if occupancy >= 90:
        print("Suggestion: Consider Opening Additional Show")


# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 seats booked

seat_occupancy(total_seats, booked_seats)
```

```python
def classify_server_load(cpu_readings):

    total = 0

    # Loop to calculate total
    for reading in cpu_readings:
        total += reading

    # Calculate average
    average = total / len(cpu_readings)

    print("Average CPU Load:", int(average), "%")

    # Classification
    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

```python
def monitor_resources(resource_usage):

    overused = []

    # Loop through dictionary
    for resource, hours in resource_usage.items():

        # Check overuse condition
        if hours > 8:
            overused.append(resource)

    # Display overused resources
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(usage_data)
```

```python
def manage_registrations(capacity, total_registrations):

    confirmed = 0
    waitlist = 0

    # Loop through each registration
    for i in range(total_registrations):

        if confirmed < capacity:
            confirmed += 1
        else:
            waitlist += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)

    if waitlist > 0:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# Sample Input
capacity = 100
registrations = 105

manage_registrations(capacity, registrations)
```
