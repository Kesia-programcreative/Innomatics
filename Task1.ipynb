```python
#Credentials
username = "admin"
password = "1234"

#Taking input
u = input("Enter username: ")
p = input("Enter password: ")

#Checking
if u == username and p == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
marks = [45, 78, 90, 33, 60]
p_count = 0
f_count = 0
for m in marks:
    if m >= 50:
        p_count += 1
    else:
        f_count += 1
print("Pass:", p_count)
print("Fail:", f_count)
```

```python
# Original list
names = [" Alice ", "bob", " CHARLIE "]

# empty list
cleaned_names = []

for name in names:
    cleaned = name.strip().lower()
    cleaned_names.append(cleaned)

# Print cleaned list
print(cleaned_names)
```

```python
message = ["Hi, Welcome to this world"]
for msg in message:
    length = len(msg)
    print("Message:", msg)
    print("Length:", length)
    if length > 10:
        print("Flagged: Long Message")
    print()
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count = 0
for log in logs:
    if log == "ERROR":
        count += 1
print("Number of ERROR entries:", count)
```
