```
emp = {
    "Ria": 85000,
    "Ann": 8000,
    "Dia": 50000
}
name = max(emp, key=emp.get)
salary = emp[name]
print(name, salary)
```

```
numbers = [40, 20, 80, 10, 60]
max_num = max(numbers)
min_num = min(numbers)
print("Max:", max_num)
print("Min:", min_num)
```

```
prices = [450, 1200, 899, 1500, 300]
count = 0
for price in prices:
    if price > 1000:
        count += 1
print("Products above 1000:", count)
```

```
attendance = ["P", "P", "A", "P"]
present_count = 0
for status in attendance:
    if status == "P":
        present_count += 1
total_days = len(attendance)
percentage = (present_count / total_days) * 100
print("Attendance Percentage:", percentage)
```

```
text = "Java"
char_count = {}
for char in text:
    char_count[char] = text.count(char)
print(char_count)
```
