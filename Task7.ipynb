```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"  # Default state

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")


light_name = "Bedroom Light"
action = "ON"

light = SmartLight(light_name)

if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()

light.display_status()
```

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


employee_name = "Rahul"
employee_id = "EMP102"
department = "IT"

emp = Employee(employee_name, employee_id, department)

emp.display_id_card()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")


contact_name = "Anita"
phone_number = "9876543210"

contact = Contact(contact_name, phone_number)

contact.display_contact()
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")


product_name = "Headphones"
price = 2499


item = Product(product_name, price)

item.print_price_tag()
```

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")


movie_name = "Inception"
rating = 4.8

film = Movie(movie_name, rating)

film.display_details()
```

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def print_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")


customer_name = "Suman"
address = "Hyderabad"

order = Delivery(customer_name, address)

order.print_delivery_details()
```
